```python
#cleansing the dataset and removing the unnecessary columns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#reading the csv file and transforming the data, droping null values and duplicates, converting to the appropriate datatypes
#this code is used for the different years of the white house dataset, the only variable that needs to be changed is the csv file name
ds16=pd.read_csv('whitehouse_waves-2016_3.csv', delimiter=",", low_memory=False)
df16=pd.DataFrame(ds16[['APPT_MADE_DATE','Total_People']])
df16=df16.drop_duplicates()
df16['APPT_MADE_DATE']=pd.to_datetime(df16['APPT_MADE_DATE'],errors='coerce')
df16=df16[pd.notnull(df16['APPT_MADE_DATE'])]
df16.dropna(subset=['APPT_MADE_DATE'])
df16['Total_People']=pd.to_numeric(df16['Total_People'],errors='coerce')
```

```python
#sorting and reseting the index of the dataset
df16.sort(columns='APPT_MADE_DATE',ascending=True)
df16=df16.reset_index(drop=True)
df16.to_csv('Cleaneddf16.csv')
```
